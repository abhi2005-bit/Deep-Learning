```python
import torch
from torch import nn
import matplotlib.pyplot as plt

torch.__version__
```

```python
#1.data preparing and loading
#we will use linear regression formula to make a straight line with *known* **parameters**

```

```python
weight = 0.7
bias = 0.3

#create
start = 0
end = 1
step = 0.02
X= torch.arange(start,end,step).unsqueeze(dim = 1)
y = weight * X +bias
X[:10],y[:10]
```

```python
#splitting data
train_split= int(0.8 * len(X))
train_split
x_train,y_train = X[:train_split],y[:train_split]
x_test,y_test=X[train_split:],y[train_split:]

len(x_train),len(y_train),len(x_test),len(y_test)
```

```python
def plot_predictions(
    train_data=x_train,
    train_labels=y_train,
    test_data=x_test,
    test_labels=y_test,
    predictions=None
):
    plt.figure(figsize=(10,7))
    plt.scatter(train_data, train_labels, c="blue", label="Training data")
    plt.scatter(test_data, test_labels, c="green", label="Testing data")
    if predictions is not None:
        plt.scatter(test_data, predictions, c="red", label="Predictions")

#shown legends
plt.legend(prop={"size":14});
```

```python
plot_predictions()
```

```python
from torch import nn
#building our first pytorch model
#createing a linear regression class model
#what our model does:
#start with random values
#look at training data and adjust the random values to better represent the ideal values we used to create the data
# it does through gradient decent and backpropagation
class LinearRegressionModel(nn.Module):
  def __init__(self):
    super().__init__()
    #Initialize model parameters
    self.weights = nn.Parameter(torch.randn(1,
                                            requires_grad=True,#the parameter can be updated via gradient descent
                                            dtype=torch.float))
    self.bias = nn.Parameter(torch.randn(1,
                                         requires_grad=True,
                                         dtype=torch.float))
  #forward meathod to define the computation in the model
  def forward(self,x: torch.Tensor)-> torch.Tensor:
    return self.weights * x + self.bias
```

```python
torch.manual_seed(42)
torch.randn(1)
```

```python
##pytorch model building essentials
#torch.nn-contains all the buildinmg blocks for computational graphs
#torch.nn.Parameters-what parameter should our model try and learn often pytorch layer from torch.nn set these for us
#torch.nn.module-base class for all neural network modules,if there is a subclass you should overwrite forward
#torch.optim-this is where the optimizers in pytorch live,they will help in gradient descent
#def forward - all nn module subclasses requires you to overwrite forward this meadthon defines what happens in forward computation
```

```python
#we check whats inside our model using .parameters()
torch.manual_seed(42)

#create an instance of model
model_0 = LinearRegressionModel()
#check the parameters
model_0.state_dict()
```

```python
weight,bias
```

```python
# making predictions using torch.inference_mode()
# to check the models predictive power ,lets see how well it predicts y_tests based on x_tests
# when we pass data through pur model it's going to run it through the forward() meathod

with torch.inference_mode():
  y_preds=model_0(x_test)

y_preds,x_test
```

```python
plot_predictions(x_test,y_preds)
```

```python
# # train mode:
# the whole idea of training is for a model to move from unknown parameters to known parameters
#one way to measure how poor or how wrong your models predictions are is to use a loss function.
#loss function is also called cost function or criterion function
```

```python
# Things we need to :
# #Loss function - is basically a function to measure how wrong our model prediction are as compared to actual value
# #optimizer - takes into account the loss of a model and adjust the models parameters(eg.weights and bias in our case )to improve the loss functions
```

```python
list(model_0.parameters())
```

```python
model_0.state_dict()
```

```python
#set up a loss function
loss_fn=nn.L1Loss()
#setup a optimizer(stochastic gradient descent)
optim_fn=torch.optim.SGD(model_0.parameters(),
                         lr=0.01)#learning rate -smaller learning rate results in small chnages and larger learning rate results in large changes
```

```python
#building a training loop
# a couple of things we need in a training loop:
# 1.Loop through the data
# 2.forward propagation
# 3.calculate the loss
# 4.optimize the zero grad
# 5.loss backward-move backwards through the network to calculate the gradients of each of the parameters of our model with respect the loss
# 6.optimizer step - use the optimizer to adjust our models parameter to try and improve the loss
```

```python
#an epoch is one loop through the data
manual_seed=42
epochs = 200
epoch_count=[]
loss_values=[]
test_loss_values=[]
##training
#loop through the data
for epochs in range(epochs):
  #set the model training loop
  model_0.train()#train mode in pytorch sets all parameters that require gradients to require gradients

  #forward pass
  y_pred = model_0(x_train)

  #calculate the loss(how different are the models predictions to the true values)
  loss = loss_fn(y_pred,y_train)


  #optimizer zero grad
  optim_fn.zero_grad()

  #loss backward
  loss.backward()

  #optimizer step
  optim_fn.step()#by default how the optimizer change will accumulate throught the loop we have to zero them above in step 3 for the next iterationof the loop

  model_0.eval()#turns off gradient tracking(turns off different settings in the model not needed for evaluation)
  with torch.inference_mode():
    test_pred = model_0(x_test)
    test_loss = loss_fn(test_pred,y_test)
  if(epochs % 10 ==0):
    epoch_count.append(epochs)
    loss_values.append(loss)
    test_loss_values.append(test_loss)
    print(f"epoch : {epochs} | loss : {loss} | test loss : {test_loss} ")
```

```python
# plot the loss corve
plt.plot(epoch_count,np.array(torch.tensor(loss_values).numpy()),label="train loss")
plt.plot(epoch_count,test_loss_values,label="test loss")
plt.title("loss curve")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend();
```

#  Putting All Together

```python
#import
import torch
import matplotlib.pyplot as plt
from torch import nn
#check python version
torch.__version__
```

```python
#Create device agnostic code
#This means if we have access to a gpu our code will use it for potentially faster computing
#If no gpu available code will default to cpu

device = "cuda " if torch.cuda.is_available() else "cpu"
print(f"device is : {device}")
```

```python
#DATA
#create some data using the linear regression formula of y=weight +X * bias
weight = 0.7
bias = 0.3

#create range values
start = 0
end = 1
step = 0.02
X=torch.arange(start,end,step).unsqueeze(dim=1)
y=weight * X +bias
X[:10],y[:10]
```

```python
train_split=int(0.8 *len(X))
x_train,y_train=X[:train_split],y[:train_split]
x_test,y_test=X[train_split:],y[train_split:]
len(x_train),len(y_train),len(x_test),len(y_test)
```

```python
def plot_predictions(
    train_data=x_train,
    train_labels=y_train,
    test_data=x_test,
    test_labels=y_test,
    predictions=None
):
    plt.figure(figsize=(10,7))
    plt.scatter(train_data, train_labels, c="blue", label="Training data")
    plt.scatter(test_data, test_labels, c="green", label="Testing data")
    if predictions is not None:
        plt.scatter(test_data, predictions, c="red", label="Predictions")

#shown legends
plt.legend(prop={"size":14});
```

```python
#plot the data
plot_predictions(x_train,y_train,x_test,y_test)
```

```python
#Building a pytorch linear model
class LinearRegressionModelV2(nn.Module):
  def __init__(self):
    super().__init__()
    #use nn.Linear() for creating the model parameters
    self.linear_layer = nn.Linear(in_features =1,
                                  out_features=1)
  def forward(self,x:torch.Tensor)-> torch.Tensor:
     return self.linear_layer(x)
#set the manual seed
torch.manual_seed(42)
model_1 = LinearRegressionModelV2()
```

```python
model_1.state_dict()
```

```python
#check the model current device
next(model_1.parameters()).device
```

```python
#set the model to use the target device
device = "cuda" if torch.cuda.is_available() else "cpu"
model_1.to(device)
next(model_1.parameters()).device
```

```python
# Training code
##setup loss function
loss_fn=nn.L1Loss()
##setup our optimizer
optimizer = torch.optim.SGD(model_1.parameters(),lr=0.001)
```

```python
#Trainig loop
torch.manual_seed(42)

epochs=500
#device agnostic code for data
x_train=x_train.to(device)
x_test=x_test.to(device)
y_train=y_train.to(device)
y_test=y_test.to(device)

#loop through the data
for epoch in range(epochs):
  model_1.train()

  #forward pass
  y_pred=model_1(x_train)

  #calculate loss
  loss=loss_fn(y_pred,y_train)

  #optimizer
  optimizer.zero_grad()

  #perform backpropagation
  loss.backward()

  #optimizer step
  optimizer.step()

  #testing
  model_1.eval()
  with torch.inference_mode():
    test_pred=model_1(x_test)
    test_loss=loss_fn(test_pred,y_test)

  #print out whats happening
  if epoch % 10 ==0:
    print(f"epoch : {epoch} | loss : {loss} | test loss : {test_loss}")
```

```python
model_1.state_dict()
```

```python
#turn the model into evaluation mode
model_1.eval()
with torch.inference_mode():
  y_preds=model_1(x_test)
y_preds
```

```python
#check model predictions visually
plot_predictions(predictions=y_preds.cpu())
```

```python
##Saving and loading our model
from pathlib import Path
#create models directory
MODEL_PATH=Path("models")
MODEL_PATH.mkdir(parents=True,exist_ok=True)

MODEL_NAME="abhijitsinha_model.pth"
MODEL_SAVE_PATH=MODEL_PATH/MODEL_NAME
MODEL_SAVE_PATH

torch.save(obj=model_1.state_dict(),
           f=MODEL_SAVE_PATH)
#load model
loader_model_1=LinearRegressionModelV2()
loader_model_1.load_state_dict(torch.load(f=MODEL_SAVE_PATH))
loader_model_1.state_dict()
```
